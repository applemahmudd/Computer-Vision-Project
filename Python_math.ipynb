**Co-factor**

```python
def cofactor_matrix(M):
  C = np.zeros(M.shape)
  n= M.shape[0]

  for i in range(n):
    for j in range(n):
      minor = np.delete(np.delete(M, i, axis=0), j, axis=1)
      C[i, j] = ((-1)**(i+j)) * np.linalg.det(minor)
      return C



  C = cofactor_matrix(A)
  print(C)
```

```python
print(cofactor_matrix(A))
```

```python
np.transpose(cofactor_matrix(A))
```

```python
invA = np.linalg.inv(A)
print(invA)
```
