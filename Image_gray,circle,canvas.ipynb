```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/GettyImages-1290176582-1738668128.jpg',0)
```

```python
cv2_imshow(img)
```

```python
cv2.imwrite('Gray image.jpg',img)
```

```python
img.shape
```

```python
resize_image = cv2.resize(img,(200,200))
```

```python
cv2_imshow(resize_image)
```

```python
rot = cv2.rotate(resize_image, cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rot)
```

```python
cv2.imwrite('Resize image.jpg',resize_image)
```

```python
crop = resize_image[30:60,40:70]
```

```python
cv2_imshow(crop)
```

```python
canvas = np.zeros((300,300,3), dtype = 'uint8')
cv2_imshow(canvas)
```

```python
img_line = canvas.copy()
```

```python
cv2.line(img_line,(50,50),(200,50),(0,0,255),3)
```

```python
img_rect = canvas.copy()
cv2.rectangle(img_rect, (50,50),(200,200),(0,0,255),3)
```

```python
img_cir = canvas.copy()
cv2.circle(img_cir, (150, 150), 100, (0, 255, 0), 3)
```
