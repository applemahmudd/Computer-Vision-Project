```python
canvas= np.zeros((300,300,3),dtype='uint8')
cv2_imshow(canvas)
```

```python
img_line=canvas.copy()
```

```python
cv2.line(img_line,(50,50),(200,50),(255,0,0),2)
```

```python
img_rect=canvas.copy()
cv2.rectangle(img_rect,(50,50),(200,200),(0,255,0),2)
```

```python
img_circle=canvas.copy()
cv2.circle(img_circle, (150, 150), 80, (0, 0, 255), 2)
```
